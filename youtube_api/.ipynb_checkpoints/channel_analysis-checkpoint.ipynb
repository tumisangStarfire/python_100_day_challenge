```python
import json
import pandas as pd
import matplotlib.pyplot as plt

channel_file = "zibo_bantsi.json"
data = None
with open(channel_file,"r") as data_file:
    data = json.load(data_file)

channel_id, stats = data.popitem() 
channel_stats = stats["channel_statistics"]
video_stats = stats["channel_video_data"]
print(channel_id)
```

```python
#view the channel statistics
print("views",channel_stats["viewCount"])
#view the channel statistics
print("subscribers",channel_stats["subscriberCount"])
#view the channel statistics
print("videos",channel_stats["videoCount"])
```

```python
#view the videos statistics
sorted_video_list = sorted(video_stats.items(),key =lambda item: int(item[1]["viewCount"]), reverse = True)
stats = []
for video in sorted_video_list :
    video_id = video[0]
    title = video[1]["title"]
    views = int(video[1]["viewCount"])
    likes = int(video[1]["likeCount"])
    #comments = int(video[1]["commentCount"])
    stats.append([title,views,likes])
    
```

```python
dataframe = pd.DataFrame(stats,columns=["title","views","likes"])
dataframe.head(10)
```

```python
dataframe = pd.DataFrame(stats,columns=["title","views","likes"])
dataframe.tail(10)
```

```python
top10 = dataframe.head(10)
axis_plot = top10.plot.bar(x="title",y="views",figsize=(12,8))
```
